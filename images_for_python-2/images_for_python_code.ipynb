```python
#importing packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
cv2.__version__
```

```python
#loading images
img1=cv2.imread("/Users/kayla/Desktop/Mat4906-ML/images_for_python-2/forest.jpg")
img2=cv2.imread("/Users/kayla/Desktop/Mat4906-ML/images_for_python-2/people-walking.jpg")
print('showing size image1', img1.shape)
print('showing size image2', img2.shape)
```

```python
#reshaping images
img1_reshaped= cv2.resize(img1,(300,300),interpolation=cv2.INTER_AREA)
img2_reshaped= cv2.resize(img2,(300,300),interpolation=cv2.INTER_AREA)
print('showing size image 2 reshaped', img2_reshaped.shape)
```

```python
#showing the images
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.imshow(img1_reshaped)
plt.subplot(222)
plt.imshow(img2_reshaped)
```

```python
#Printing RGB channels 
plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(img1_reshaped[:,:,0])
plt.gca().set_title('red channel')
plt.subplot(132)
plt.imshow(img1_reshaped[:,:,1])
plt.gca().set_title('green channel')
plt.subplot(133)
plt.imshow(img1_reshaped[:,:,2])
plt.gca().set_title('blue channel')

```

```python
#combing pictures using addition 
dst=cv2.addWeighted(img1_reshaped,0.2,img2_reshaped,0.5,0.1)
plt.figure(figsize=(10,10))
plt.imshow(dst)
```
